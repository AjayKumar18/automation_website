```python
from selenium import webdriver
import time
from dateparser.search import search_dates
import pandas as pd
from datetime import datetime
import warnings
```

```python
def fetch_result():
    try:
        thres_date=datetime(2019,6,25)
        inc=0
        final_only_dates=[]
        demo_div = driver.find_element_by_id("results")
        temp=search_dates(demo_div.get_attribute('textContent'))
        print(len(temp))
        df_for_dates=pd.DataFrame(temp,columns=["Dates","Etc"])
        df_for_dates
        only_dates=df_for_dates.Dates
        for q in range(0,len(only_dates)):
            if len(only_dates[q])==10:
                final_only_dates.append(only_dates[q])
        print(final_only_dates)        
        date_series = pd.to_datetime(final_only_dates,format='%d/%m/%Y')
        for j in range(0,len(date_series)):
            if date_series[j]>thres_date:
                print(date_series[j],thres_date,"True")
                inc=inc+1
        return(inc)
    except:
        Text=driver.find_element_by_xpath("//*[@id='msg_overlay']/h2")
        time.sleep(0.5)
        if Text.text=="Error":
            driver.find_element_by_xpath("//*[@id='msgboxclose']").click()
            print("0")
            return(0)
            
```

```python


df=pd.read_excel("MCA LINKS.xlsx");
#read the excel file

#drop the NAN values from the CIN
final_set=df.dropna(subset=["CIN/ FCRN"]).copy()
#set the index to name
final_set.set_index("Name",inplace=True)
#Fetch the CIN number from the final_set
i=0
```

```python
break_it=0
for i in range(400,420):
    
    try:
        print_var=[]
        driver.get('http://www.mca.gov.in/mcafoportal/viewPublicDocumentsFilter.do');
        driver.find_element_by_id("cinChk").click()
        driver.find_element_by_id("cinFDetails").click()
        CIN=final_set.iloc[i][1]
        #if the compnay is not lpp then it enters into the condition    
        if len(CIN)>=19:
            CIN = driver.find_element_by_id("cinFDetails").send_keys(CIN)
            driver.find_element_by_id("viewDocuments_0").click()
            driver.find_element_by_xpath("//*[@id='results']/tbody/tr[2]/td[2]/a").click()
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_finacialYear']/option[15]").click()
            #This one is for certificates(4)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[2]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()        
            final_set.certificates[i]=fetch_result()
            time.sleep(1)
            #This one is for change in directories(5)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[3]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.directories[i]=fetch_result()
            #For incorporation documents(6)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[4]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.incorporationdocuments[i]=fetch_result()
            #this one is for charge documents(7)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[5]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.chargedocuments[i]=fetch_result()
            #Annual Returns and Balance Sheet eForms(8)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[6]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.ReturnsandBalance[i]=fetch_result()
            #LLP Forms(Conversion of company to LLP)(9)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[7]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.LLPForms[i]=fetch_result()
            #Other eForm Documents(10)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[8]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.OthereForm[i]=fetch_result()
            #Other  Documents(10)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_categoryName']/option[9]").click()
            time.sleep(1)
            driver.find_element_by_xpath("//*[@id='viewCategoryDetails_0']").click()
            final_set.Other_attachment[i]=fetch_result()
            #Return to main website
    except:
        print("error on "+str(i+1))
        break_it=break_it+1
        if break_it==10:
            break
```
