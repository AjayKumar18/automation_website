```python
df=pd.read_csv("LinkedIN (465 companies).csv")
df.head()
df1=df.URL_Main
```

```python
#len(df1)
for UR in range(431,442):
    try:
        time.sleep(2)
        url=df1[UR]
        driver.get(url+'/jobs/');
        time.sleep(2)
        i=1

        #if there is job
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 



        try:
            if driver.find_element_by_css_selector("div.org-empty-jobs-module").text=="There are no jobs right now.\nCreate a job alert and we’ll let you know when relevant jobs are posted.":
                print("0")
                df.Jobs[UR]=0
                df.Total[UR]=0


        except:
            try:
                driver.find_element_by_xpath("//*[@class='link-without-hover-visited mt5 ember-view']//*[text()='See all jobs']").click()
                time.sleep(3)            
                driver.find_element_by_xpath("//*[@id='sort-by-select-trigger']/li-icon").click()            
                driver.find_element_by_xpath("//*[@id='sort-by-select-options']/li[2]/button").click()
                time.sleep(2)
                toal_pages=0
                total_p=driver.find_elements_by_class_name("artdeco-pagination__indicator")
                total_pages=len(total_p)
            except:
                total_pages=0
                total_pages=1

            #if driver.find_element_by_css_selector("div.org-empty-jobs-module").text=="There are no jobs right now.\nCreate a job alert and we’ll let you know when relevant jobs are posted.":
                #df.Jobs[UR]=0

        #Main program
            if total_pages>=0:
                thres_date=datetime(2019,6,25)
                inc=0
                total_jobs=0
                date_of_each=[]
                temp=[]
                f = "%Y-%m-%d"
                date_of_each=[]

                driver.find_element_by_css_selector("div.jobs-search-two-pane__results").click()
                for x in range(0,15):
                    time.sleep(0.5)
                    driver.find_element_by_css_selector("div.jobs-search-results").send_keys(Keys.PAGE_DOWN)
                time.sleep(0.5)
                total_jobs=driver.find_elements_by_css_selector("time.job-card-search__time-badge ")
                for l in range(0,len(total_jobs)): 
                    temp=total_jobs[l].get_attribute("datetime")
                    date_of_each.append(temp)

            if total_pages>1:
                while(True):

                    time.sleep(0.5)
                    i=i+1
                    driver.find_element_by_xpath("//button[@aria-label='Page "+str(i)+"']").click()
                    driver.find_element_by_css_selector("div.jobs-search-two-pane__results").click()
                    for x in range(0,15):
                        driver.find_element_by_css_selector("div.jobs-search-results").send_keys(Keys.PAGE_DOWN)
                        time.sleep(0.5)
                    total_jobs=driver.find_elements_by_css_selector("time.job-card-search__time-badge ")
                    for l in range(0,len(total_jobs)): 
                        temp=total_jobs[l].get_attribute("datetime")
                        date_of_each.append(temp)
                    print(date_of_each)
                    if total_pages==i:
                        break;
            for q in range(0,len(date_of_each)):
                out = datetime.strptime(date_of_each[q], f)
                if out>thres_date:
                    inc=inc+1
            print(len(date_of_each))
            print(inc)
            df.Jobs[UR]=inc
            df.Total[UR]=(len(date_of_each))
    except:
        df.Jobs[UR]="error"
        print("error")
```
