```python
#function to fetch all the video html code which are relevant
def get_channel_videos(channel_id):
    req=youtube.channels().list(id=channel_id,part='contentDetails').execute()
    playlist_id=req['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    videos=[]
    breaker=0
    i=0
    next_page_token=None
    while 1:
        req=youtube.playlistItems().list(playlistId=playlist_id,part='snippet',maxResults=50,pageToken=next_page_token).execute()
        videos+=req['items']
        next_page_token=req.get('nextPageToken')
        for i in range(0,len(videos)):
            if videos[i]['snippet']['publishedAt']<date:
                breaker=1
                break
                
        if next_page_token is None or breaker==1:
            break
            
    return videos
```

```python
#return the relevant count of the views and number of videos
def count_everything(videos):
    inc=0
    final_view_count=0
    for i in range(0,len(videos)):
        
        view=youtube.videos().list(id=videos[i]['snippet']['resourceId']['videoId'],part='statistics').execute()
        view_count=int(view['items'][0]['statistics']['viewCount'])
        if videos[i]['snippet']['publishedAt']>=date:
            final_view_count+=view_count
            inc+=1
    print(videos[0]['snippet']['channelTitle'])
    print(inc)
    print(final_view_count)
    return([inc,final_view_count,videos[0]['snippet']['channelTitle']])
```
