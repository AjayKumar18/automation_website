```python
from selenium import webdriver
import time
from dateparser.search import search_dates
import pandas as pd
from datetime import datetime
import warnings
import numpy as np
```

```python
warnings.simplefilter('ignore')
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome('C:/Users/chauhan/Desktop/python/Chrome/1')  # Optional argument, if not specified will search path.
```

```python
df=pd.read_csv("Alexa.csv");
final_set=df.dropna(subset=["URLS"]).copy()
df1=pd.DataFrame(columns = ['URL', 'Global_ranking','Local_ranking','sites_linking','Daily_time'])
df1.URL=final_set.URLS
#final_set.set_index("Name",inplace=True)
```

```python
for i in range(0,15):
    url=df1.URL[i]
    driver.get(url);
    time.sleep(15)
    global_rank=0
    daily_time=0
    local_rank=0
    site_link=0
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
    temp=driver.find_elements_by_class_name("rankmini-rank")
    try:
        global_rank=temp[0].text[1:]
    except:
        global_rank='NA'
    try:
        daily_time=temp[1].text
    except:
        daily_time='NA'

    try:
        local_temp=driver.find_elements_by_css_selector(".CountryRank")
        local_rank=local_temp[0].text[2:]
    except:
        local_rank='NA'
    try:
        sites_link_temp=driver.find_elements_by_css_selector("span.big")
        site_link=sites_link_temp[0].text
    except:
        site_link='NA'
    print(global_rank)
    print(daily_time)
    print(local_rank)
    print(site_link)
    print("\n")
    df1.Global_ranking[i]=global_rank.replace(',', '')
    df1.Local_ranking[i]=local_rank.replace(',', '')
    df1.sites_linking[i]=site_link
    df1.Daily_time[i]=daily_time
```
