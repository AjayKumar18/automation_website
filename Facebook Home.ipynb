```python
url_list =pd.read_csv(r'C:\Users\Ankita Yadav\Desktop\lightspeed\input\Facebook_links(465 companies).csv')
url_list = url_list.URLS
url_list1 = url_list.dropna()
#url_list2 = url_list1[1:]
url_list1
```

```python
import sys
```

```python
df8 = pd.DataFrame()
```

```python
for url in url_list1:
    driver.get(url)
    time.sleep(4)
    
    
    try:
        driver.implicitly_wait(1)
        element = driver.find_element_by_class_name("_2w0a").text
        element = element.split()
        Rating = element[0]
    except (NoSuchElementException,IndexError) as exception:
        Rating = "NA"

    try:
        element = driver.find_element_by_class_name("_2w0b").text
        element = element.split()
        Reviews = element[5]
        driver.implicitly_wait(1)
    except (NoSuchElementException,IndexError) as exception:
        Reviews = "NA"
        
    try:
        element = driver.find_element_by_class_name("_6590").text
        element = element.splitlines()
        Follows = element[4]
        driver.implicitly_wait(2)
    except (NoSuchElementException,IndexError) as exception:
        Follows="NA"
    
    try:
        element = driver.find_element_by_class_name("_6590").text
        element = element.splitlines()
        Likes = element[3]
        driver.implicitly_wait(2)
    except (NoSuchElementException,IndexError) as exception:
        Likes = "NA"
        
        
        
            
    if sys.version_info[0] < 3: 
        
        from StringIO import StringIO
    else:
        from io import StringIO

    #TESTDATA = pd.DataFrame(columns=['Website','Global_Rank','India_Rank','Duration_on_site','Sites_linking_in'])
    TESTDATA=StringIO("""Rating;Reviews;Follows;Likes;URLS""")

    df = pd.read_csv(TESTDATA, sep=";")


    df.Rating = [Rating]
    df.Reviews = [Reviews]
    df.Follows = [Follows]
    df.Likes = [Likes]
    df.URLS = [url]
    #df.Website = [Website]
    df8=df8.append(df)
```
